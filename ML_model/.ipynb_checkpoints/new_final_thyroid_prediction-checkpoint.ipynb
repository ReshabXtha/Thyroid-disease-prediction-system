### Correlation between data

```python
target = 'Class'
x = df.loc[:, df.columns != target]
y = df.loc[:, df.columns == target]
print(f'shape of x = {x.shape}, shape of y={ y.shape}')
```

```python
x_plot = ['Normal', 'Hypothyroidism', 'Hyperhyroidism']
y_plot = [166, 367, 6665]
plt.bar(x_plot, y_plot, data=df, color='blue')
plt.xlabel('Classes')
plt.ylabel('value count')
plt.title('Nuber of data present with respect to classes')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
cormat = df.corr()
```

```python
# sns.pairplot(final_dataset)
top_corr_feature = cormat.index
plt.figure(figsize=(20, 20))
g = sns.heatmap(df[top_corr_feature].corr(), annot=True, cmap='RdYlGn')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
classifier = RandomForestClassifier(n_estimators=150, random_state=0, n_jobs=-1)
classifier_cv = RFECV(estimator=classifier, step=1, cv=5, scoring='accuracy')
model = classifier_cv.fit(x_train, np.ravel(y_train,order='C'))
print('Optimal number of features :', classifier_cv.n_features_)
print('Best features :', x_train.columns[classifier_cv.support_])
```
